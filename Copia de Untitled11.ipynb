#Como importar librerias y dataset a colab
**Conceptos**

**Libreria**: también conocida como biblioteca es un conjunto de funciones, clases, métodos u otros recursos que están agrupados en un solo archivo o conjunto de archivos. Estas funciones y recursos están diseñados para ser reutilizados en varios programas sin necesidad de escribir el código desde cero.

**Dataset**: Conjunto de datos es una colección de datos organizados que pueden ser utilizados para diversos fines, como el análisis, entrenamiento de modelos de machine learning, o simplemente almacenamiento y manipulación de información.

```python
###Muestreo aleatorio sistematico
###seleccion sistematica de la muestra, sistema aleatorio esta es nuestra propia forma de obtener valores.
```

```python
### Definicion del intervalo k (tamaño de la poblacion / el tamaño de la muestra deseada)
```

Definir tamaño muestral:

```python
sample_size = 100
interval_k = population_size//sample_size
```

muestra aleatoria simple: Seleciona un conjunto representatico numeros al alzar utilizando metodos de programacion

hilo es un metodo de arreglo y se llama a travez de corchetes, selciona la informacion dentro de lo que vemos y la forma mas sencilla de escribir todo este arreglo es:

```python
systematic_sample = data.iloc[::interval_k]
```

```python
print("muestra obtenida por muestreo sistematico")
display(systematic_sample)
```

agrupar la poblacion por idioma de programacion

```python
grouped_by_dev_leguage = data.groupby("Idioma de programación")
grouped_by_dev_leguage.head()
```

Con analisis y sacando la media podemos sacar valores.

```python
group_analysis = grouped_by_dev_leguage[["Líneas de código", "Número de bugs"]].mean()
print("analisis por lenguaje de programacion")
display(group_analysis)
```
